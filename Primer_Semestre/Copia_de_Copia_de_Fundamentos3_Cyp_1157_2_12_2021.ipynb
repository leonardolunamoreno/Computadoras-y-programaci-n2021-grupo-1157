#Estructura de control de repetici칩n Haz - mientras 

Una estructura de control haz -mientras nos permite ejecutar repetidamente mientras una condici칩n se siga cumpliendo 

# sintaxis 

while<condici칩n> 
    <Bloque de sentencias a ejecutar repetidamente> 
<siguiente sentencia posterior a while>

La diferencia principal en comparaci칩n con la estructura for es que con while no conoces el n칰mero de ciclos a ejecutar, en for si se conoce. 
C칩mo ejemplo iniciemos con el siguiente enunciado. 
Hacer un programa que solicite un valor entero positivo y calcule repetidamente la mitad y se reasigne a la misma variable. El ciclo se debe detener cuando el resultado sea menor que 1. 
mostrar en pantalla el resultado de cada c치lculo y al final el n칰mero de iteraciones que realiz칩. 

```python
valor = int(input("Introduce un n칰mero entero positivo: "))
contador = 0 
if valor > 0:
    while valor > 1.0: 
        valor = valor /2 
        print(round(valor,2))
        contador +=1 
else: 
    print("El n칰mero es negativo, intenta de nuevo")
print("Total de iteraciones=", contador)
```
