Este código mostrado a continuación permite comprobar que un micrófono conectado a una computadora o laptop, efectivamente puede grabar audio, aunque también es posible comprobar que el propio microfóno incorporado en la computadora o laptop funcionan.

```python
# Comprobar que un micróno está grabando audio en caso de estar conectado
import sounddevice as sd
import numpy as np

duracion = 1.0
fs = 44100

print("Grabando...")
audio = sd.rec(int(duracion * fs), samplerate=fs, channels=1, dtype='float32')
sd.wait()

señal = audio.flatten()

print("Primeros 50 valores de la señal:")
print(señal[:50000])

print("Valor máximo detectado:", np.max(señal))
print("Valor mínimo detectado:", np.min(señal))


## import sounddevice as sd
#import pprint

#pprint.pprint(sd.query_devices())
# print("Default:", sd.default.device)
```

Este código permite mapear cada puerto relacionado con el audio de una computadora, ya sean micrófonos o bocinas.

```python
import sounddevice as sd
import pprint

# Lista todos los dispositivos reconocidos
devices = sd.query_devices()
pprint.pprint(devices)

# Muestra cuál es el dispositivo por defecto
print("Dispositivo por defecto (in, out):", sd.default.device)

# Lista índices de dispositivos que tienen capacidad de entrada
input_indices = [i for i, d in enumerate(devices) if d['max_input_channels'] > 0]
print("Dispositivos con entrada (índices):", input_indices)

# Si hay al menos uno, muéstralo y opcionalmente configúralo como default
if input_indices:
    idx = input_indices[0]
    print("Usando este dispositivo de entrada por índice:", idx)
    print(devices[idx])
    sd.default.device = idx  # fuerza uso de ese dispositivo
else:
    print("No se detectaron dispositivos de entrada (micrófono).")
```
